```python
# Take the state in q0 and teleport in q2
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector
```

```python
circuit = QuantumCircuit(3,3) #3 qbits and 3 classicalbit
```

```python
%matplotlib inline
circuit.draw(output="mpl")
```

```python
#teleporting state 1
#circuit.x(0)
#teleporting (|0> + |1>)/sqrt(2) i.e Hadamard state
circuit.h(0)
circuit.barrier()
circuit.draw(output="mpl")
```

```python
#create Entanglement from q1 tp q2
circuit.h(1) #Hadamard gate to q1
circuit.cx(1,2)
```

```python
circuit.draw(output="mpl")
```

```python
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output="mpl")
```

```python
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output="mpl")
```

```python
circuit.barrier()
circuit.measure(2,2)
circuit.draw(output="mpl")
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots= 1024).result()
counts= result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
